# Network models

Nell'ambito della letteratura psicometrica recente, è emersa un'alternativa innovativa per la rappresentazione dei costrutti psicometrici. Lo scopo di questo capitolo è introdurre e illustrare alcuni concetti relativi ai modelli di reti applicati alla psicometria.

## Network psychometrics

Invece di considerare le relazioni tra le variabili osservate come derivate da una variabile latente sottostante correlata a tutte le variabili osservate, si adotta l'idea di concepire il gruppo di variabili osservate come una rete in cui interazioni causali collegano tali variabili. Questo nuovo approccio rientra nel campo della "network psychometrics" (Marsman et al. 2018; Epskamp et al. 2018) e mira a estrarre modelli di rete da dati psicologici. Lo scopo è di mappare le complesse interazioni tra componenti psicologiche, biologiche, sociologiche e altre.

All'interno di questi modelli di rete, le variabili osservate sono considerate nodi, mentre la forza dell'associazione condizionale tra due variabili, controllando tutte le altre variabili, viene rappresentata come archi. Parametri di un livello superiore, che possono rappresentare variabili latenti associate a persone e situazioni, influenzano le connessioni previste nella rete. In certi casi, le equazioni per le distribuzioni congiunte di probabilità delle variabili osservabili, date i parametri sottostanti in un modello di rete, risultano essere analoghe a quelle derivanti da un modello di teoria di risposta agli item (IRT) con variabili relative alla persona e al compito. Questi modelli combinano sviluppi psicometrici, ovvero i modelli IRT e l'analisi delle reti sociali.

Per esemplificare, consideriamo gli elementi seguenti (adattati dalla scala di stabilità emotiva IPIP50):

- Item 1: "Mi sento in preda al panico facilmente."
- Item 2: "Le emozioni mi travolgono spesso."
- Item 3: "Mi lamento frequentemente delle cose."

L'analisi fattoriale ha dimostrato che questi (e altri) item misurano tutti lo stesso tratto; questi item sono correlati tra loro e quindi formano un fattore. Tuttavia, le correlazioni non implicano causalità e quindi i modelli di analisi fattoriale non possono evidenziare le relazioni di causalità tra gli item.

Ad esempio, se qualcuno è fortemente d'accordo nel dire che si sente in preda al panico facilmente, logicamente dirà anche che le emozioni lo travolgono spesso. Il panico implica una risposta emotiva travolgente, quindi il modo in cui una persona risponde all'item 1 dovrebbe influenzare causalmente anche la risposta all'item 2. Tuttavia, il collegamento inverso è probabilmente meno forte. Solo poche persone che concordano nel sentirsi travolti dalle emozioni entreranno in panico. Potrebbero invece sentirsi depressi. L'item 3 probabilmente non ha un collegamento causale significativo né con l'item 1 né con l'item 2; provare panico o sentirsi travolti emotivamente non implica lamentele, né è evidente perché lamentarsi dovrebbe implicare sentimenti di panico o essere travolti dalle emozioni.

Calcolando correlazioni parziali, l'analisi di rete dovrebbe essere in grado di identificare tali effetti, cosa che l'analisi fattoriale non può fare. Tuttavia, la domanda cruciale è se l'analisi di rete e l'analisi fattoriale producano risultati molto diversi quando applicate allo stesso insieme di dati.

## Limiti della network analysis

Esaminando le idee espresse, risulta pratica comune trattare le ridotte correlazioni parziali come se fossero nulle durante l'analisi delle reti. Ciò semplifica il diagramma, poiché le connessioni ritenute insignificanti possono essere omesse dalla rappresentazione grafica, evitando così che ogni nodo sia connesso a tutti gli altri nodi. Tale procedura è ampiamente eseguita attraverso la tecnica del "lasso" (least absolute shrinkage and selection operator). Qui si effettua una scelta in merito al valore di un parametro (lambda) e, tanto maggiore è il valore di lambda selezionato, tanto più archi il procedimento lasso forza a diventare nulli. Questo processo tecnico è noto come "render il modello più sparso".

Questo principio è analogo a trattare i piccoli carichi fattoriali come se fossero nulli durante l'interpretazione delle matrici di fattori rotati. Tuttavia, questo è altrettanto arbitrario, sia nella scelta del metodo che del valore di lambda utilizzato. Tale aspetto è motivo di preoccupazione poiché sorge la questione se la stessa struttura emerga quando lo stesso ricercatore analizza due campioni di dati provenienti dalla stessa popolazione.

Un secondo problema riguarda la possibilità che un nodo possa essere influenzato da un elevato numero di altri nodi, ma ciascuna singola relazione sia di dimensioni ridotte e possa essere considerata nulla quando si rende la rete più sparsa. Un esempio chiaro si riferisce all'analisi delle influenze genetiche sull'intelligenza. Benché sia noto che l'intelligenza sia ampiamente influenzata dalla predisposizione genetica delle persone, ciò coinvolge un vasto numero di geni (altri nodi), ciascuno dei quali apporta un'influenza impercettibile. Individualmente, il loro contributo sarebbe minimo (e alcuni potrebbero facilmente essere posti a zero), ma considerati nel loro insieme, l'effetto può diventare significativo.

Inoltre, rimane incerta la dimensione dei campioni necessaria. Ci sono regole empiriche: ad esempio, Epskamp et al. (2017) suggeriscono che il numero dei campioni non dovrebbe essere inferiore a n(n − 1), dove n è il numero dei nodi. Tuttavia, questi autori dimostrano che tale regola potrebbe ancora portare a stime instabili. In ogni caso, è probabile che la dimensione del campione necessaria sia correlata alla quantità di errore di misurazione associata a ciascuna variabile. Benché tali problemi siano stati trattati nell'ambito dell'analisi fattoriale, non sembra che siano stati esaminati dettagliatamente dagli analisti di reti.

Un quarto problema riguarda la varietà di algoritmi sviluppati per rappresentare la struttura dei dati, quindi l'aspetto del grafico dipenderà in parte dalla scelta arbitraria compiuta. I diagrammi di rete possono apparire accattivanti, ma è essenziale studiare attentamente i dettagli di ciò che rappresentano in modo preciso.

## Modelli direzionali

Fino a questo punto, abbiamo esaminato i modelli di analisi delle reti basati su correlazioni, i quali prevedono al massimo un singolo collegamento tra ciascuna coppia di nodi. Questa limitazione deriva dal fatto che l'analisi parte da una matrice di correlazione, e all'interno del diagramma di rete, gli archi rappresentano correlazioni parziali: ovvero la forza della relazione tra due nodi dopo che tutte le altre correlazioni sono state corrette. Tuttavia, va tenuto presente che le interazioni tra i nodi non sono sempre simmetriche.

Per esemplificare, consideriamo il caso in cui il quoziente di intelligenza (QI), misurato attraverso test di intelligenza, sia strettamente correlato al numero di anni di istruzione ricevuti da un individuo. Questa correlazione solleva una domanda: quale potrebbe essere il motivo sottostante? Una possibile spiegazione è che l'intelligenza possa influenzare il rendimento scolastico, in modo tale che individui con un QI elevato ottengano risultati migliori, ricevano rinforzi positivi attraverso voti alti e quindi scelgano di continuare gli studi il più a lungo possibile. Al contrario, è altrettanto possibile che frequentare la scuola possa influenzare l'intelligenza stimolando lo sviluppo del pensiero astratto e delle abilità linguistiche, tra gli altri fattori che migliorano le performance nei test di intelligenza. In questo scenario, un modello direzionale di rete che includa il QI e gli anni di istruzione come nodi potrebbe mostrare due collegamenti tra questi nodi, evidenziando collegamenti causali (con probabili dimensioni diverse) in entrambe le direzioni. Questo stesso principio può essere applicato anche agli elementi dei test.

Un approccio alternativo consiste nell'utilizzare una matrice di similarità come base per un modello di rete. Questo approccio è spesso adottato quando si tratta di mappare reti sociali, utilizzando le persone come nodi. Ad esempio, consideriamo due individui, Claire e Ryan: mentre Claire potrebbe considerare Ryan un amico stretto, Ryan potrebbe invece percepire Claire come una semplice conoscenza. Questo esempio dimostra come una matrice di "similarità" possa rappresentare le relazioni in modo più flessibile rispetto alle tradizionali matrici di correlazione.

## Un esempio concreto

Consideriamo una rete sociale basata derivata da dati che forniscono informazioni relazionali, in cui delle persone interagiscono tra loro in qualche modo. Come esempio didattico, consideriamo quattro persone: Horst, Helga, Klaus e Gertrud. Nelle reti, ciascuno dei quattro attori sarà rappresentato da un nodo (o vertice). In questo esempio, le interazioni avvengono quando gli attori hanno un qualche scambio materiale (qui chiamato album). Supponiamo che Helga e Horst abbiano scambiato alcuni oggetti. In una rete, sono quindi collegati da un edge (o arco). Helga ha dato a Horst tre oggetti, mentre Horst ha dato due oggetti ad Helga. Abbiamo un edge Helga → Horst e un edge Horst → Helga. Il primo edge ha un peso di 3, e il secondo edge ha un peso di 2. L'intero insieme di scambi di oggetti può essere rappresentato da un elenco di edges. Utilizzando il pacchetto igraph (Csárdi e Nepusz, 2006; Kolaczyk e Csárdi, 2014), creiamo l'elenco corrispondente di edges e assegnamo i pesi agli edges come segue:

```r
library("igraph")
album_df <- matrix(c(
    "Helga", "Gertrud", "Horst", "Klaus",
    "Horst", "Helga", "Gertrud", "Horst", "Klaus", "Horst",
    "Horst", "Helga", "Helga", "Klaus", "Klaus", "Klaus",
    "Gertrud", "Gertrud"
), ncol = 2)
album_el <- graph.edgelist(album_df, directed = TRUE)
E(album_el)$weight <- c(2, 3, 3, 4, 5, 1, 7, 7, 2)
```

Un secondo modo equivalente di rappresentare i dati di una rete è attraverso una matrice di adiacenza, in cui le righe riflettono il "da" e le colonne riflettono il "a".

```r
album_ad <- get.adjacency(album_el,
    sparse = FALSE,
    attr = "weight"
)
album_ad
```

Il grafo di rete corrispondente può essere prodotto nel seguente modo:

```r
plot(album_el,
    vertex.size = 0, edge.arrow.size = 0.5,
    vertex.label.dist = 0.8, vertex.color = "black",
    vertex.label.color = "black", vertex.label.cex = 1.5
)
```

Il diagramma presente nella figura rappresenta una rete diretta e ponderata, in quanto gli edge esprimono una relazione "da-a", e i loro valori riflettono l'intensità della connessione. Sarebbe possibile trasformarlo agevolmente in una rete non ponderata sostituendo ciascun valore degli edge maggiore di 0 con 1 (cioè rappresentando uno scambio o l'assenza di scambio). Altresì, potremmo convertirlo in una rete non diretta in cui abbandoniamo la distinzione "da-a", ottenendo la somma degli elementi corrispondenti nella parte triangolare superiore e inferiore della matrice di adiacenza, la quale diventerebbe simmetrica. Come si può osservare, emergono quattro tipi fondamentali di reti basati sulla combinazione di edge diretti/non diretti e pesati/non pesati.

### Reti di correlazione

In psicologia, la raccolta di dati in forma relazionale è rara. Al contrario, si fa uso delle reti di correlazione, le quali si basano su una matrice di correlazione. Queste reti sono ampiamente utilizzate nell'ambito clinico, dove trovano applicazione, ad esempio, nella rilevazione di associazioni tra sintomi.

In questa sezione illustreremo le reti di correlazione attraverso un insieme di dati proveniente da McNally et al. (2017). Questo dataset sulla comorbidità tra disturbo ossessivo-compulsivo (DOC) e depressione comprende 16 elementi relativi alla depressione (QIDS; scala a 5 punti) e 10 elementi relativi al disturbo DOC (Y-BOCS; scala a 4 punti). Il campione conta n = 408 pazienti. Come primo passo, procediamo al calcolo della matrice di correlazione R:

```r
library("MPsychoR")

data(Rogers)
cormat <- cor(Rogers)
cormat |> round(2) |> print()
```

Un problema che sorge dall'utilizzo di matrici di correlazione all'interno di un contesto di rete è che spesso portano a un grafo completamente connesso, poiché le correlazioni empiriche in generale non sono pari a 0. Approcci semplici per affrontare questa problematica includono l'eliminazione delle correlazioni basse attraverso la scelta di una soglia di correlazione o il mantenimento solamente delle correlazioni significative (dopo una correzione adeguata dei valori p dovuta al problema dei test multipli). Nel grafico di rete presente nella Figura seguente, generato tramite l'utilizzo del pacchetto `qgraph`, abbiamo impostato la soglia minima di correlazione assoluta a 0.2 (ovvero una dimensione dell'effetto di entità piccola). Più spesso è un edge, maggiore è la correlazione. Gli edge con correlazioni negative sono colorati in modo differente.

```r
library("qgraph")
cornet <- qgraph(cormat, layout = "spring", minimum = 0.2,
graph = "cor", groups = list(Depression = 1:16, OCD = 17:26),
color = c("white", "gray"), labels = colnames(Rogers),
title = "Depression/OCD Correlation Network")
```

La figura mostra una rete di correlazione per gli item relativi alla depressione e al disturbo ossessivo-compulsivo (DOC). Le linee verdi indicano correlazioni positive, mentre le linee rosse indicano correlazioni negative.

Il pacchetto `qgraph` offre diverse opzioni, implementate nella funzione di centralità, per analizzare l'importanza e l'influenza dei nodi all'interno di una rete. La centralità è un concetto fondamentale nell'analisi delle reti ed è utilizzata per valutare il grado di importanza di ciascun nodo in base alle sue connessioni e al suo ruolo nella rete. Le principali misure di centralità incluse sono:

- **Centralità di grado (strength)**: Questa misura riflette il numero di lati (archi) connessi a ciascun nodo. I nodi con un alto grado di centralità hanno molte connessioni e possono essere considerati "hub" all'interno della rete.

- **Centralità di vicinanza**: La centralità di vicinanza misura quanto è vicino un nodo a tutti gli altri nodi nella rete, basandosi sulla lunghezza dei percorsi più brevi tra i nodi. I nodi con una centralità di vicinanza elevata possono essere considerati "centrati" nella rete, poiché possono comunicare rapidamente con altri nodi.

- **Centralità di intermediazione**: Questa misura valuta il ruolo di un nodo nel collegamento di altri nodi. I nodi con alta centralità di intermediazione si trovano spesso su percorsi di comunicazione critici tra altri nodi. In sostanza, agiscono da intermediari che controllano il flusso di informazioni o l'interazione tra altri nodi.

Un grafico rappresentativo delle misure di centralità può essere creato per visualizzare in modo chiaro e intuitivo l'importanza relativa dei nodi nella rete.

```r
centralityPlot(cornet)
```

Per espandere questa semplice rete di correlazione, possiamo adottare le seguenti modifiche. In primo luogo, anziché utilizzare le tradizionali correlazioni di Pearson, optiamo per le correlazioni parziali. Una correlazione parziale coinvolge due nodi e considera l'influenza di tutti gli altri nodi presenti nella rete. Di conseguenza, gli archi inferiti riflettono in modo più accurato l'influenza diretta tra i nodi. In secondo luogo, invece di impiegare una soglia di correlazione arbitraria o un livello di significatività corretto con il metodo di Bonferroni, possiamo considerare l'approccio del "lasso grafico" (Friedman et al., 2008).

Il pacchetto qgraph fornisce un'implementazione ad alto livello semplice da utilizzare per calcolare le correlazioni parziali regolarizzate. Il calcolo del "lasso grafico" in questo modo può essere eseguito come segue:

```r
qgraph(cormat,
    layout = "spring", sampleSize = nrow(Rogers),
    graph = "glasso",
    groups = list(Depression = 1:16, OCD = 17:26),
    color = c("white", "gray"), labels = colnames(Rogers),
    title = "Depression/OCD Graphical Lasso"
)
```

L'immagine illustra una rete di lasso grafico di correlazioni parziali che coinvolge gli elementi legati alla depressione e al disturbo ossessivo-compulsivo (DOC). Questo tipo di rete visualizza le relazioni tra gli elementi in modo più preciso, poiché le correlazioni parziali tengono conto dell'influenza degli altri nodi presenti nella rete. Le linee verdi rappresentano correlazioni parziali positive, suggerendo connessioni positive tra gli elementi. D'altra parte, le linee rosse indicano correlazioni parziali negative, evidenziando connessioni negative tra gli elementi.

Un aspetto da considerare è che la distanza tra i nodi nei grafici di rete, come quelli mostrati nelle figure precedenti, non ha un significato intrinseco. I nodi sono disposti in modo da ottenere un'organizzazione esteticamente gradevole del grafico, e quindi la loro posizione non riflette una misura di similarità o distanza. In questo contesto, approcci di scalatura come il Multidimensional Scaling (MDS) sono particolarmente utili, poiché posizionano i nodi in base alla loro somiglianza. In altre parole, i nodi con correlazioni più elevate vengono posizionati più vicini l'uno all'altro. Questo approccio fornisce una rappresentazione visiva intuitiva delle relazioni tra gli elementi.

Attraverso questa metodologia, le posizioni dei nodi risultanti possono essere ulteriormente oggetto di interpretazione. Ad esempio, i nodi che sono vicini tra loro potrebbero rappresentare elementi correlati in modo significativo, mentre quelli distanti potrebbero indicare elementi con correlazioni più deboli. Pertanto, oltre alla rappresentazione visiva, questa tecnica offre una base per ulteriori analisi e intuizioni nel contesto delle reti di correlazione.

## Commenti e considerazioni conclusive

Va sottolineato che l'analisi delle reti può richiedere un approccio iterativo e impegnativo dal punto di vista computazionale. La scelta dell'algoritmo adeguato e dei parametri appropriati può avere un impatto significativo sui risultati ottenuti. In aggiunta, la selezione del modello definitivo può presentare una certa dose di arbitrarietà, poiché diversi modelli potrebbero adattarsi in modo simile ai dati osservati. Nonostante le sfide coinvolte in questo processo, l'analisi delle reti direzionali si presenta come un campo altamente promettente. Questo campo può aprire nuove prospettive nel modo in cui comprendiamo le intricate interazioni tra variabili complesse. Nel suo insieme, l'analisi delle reti offre la possibilità di rivelare strutture nascoste e dinamiche sottostanti che possono contribuire in modo significativo al nostro approccio alla comprensione dei fenomeni complessi nei contesti psicometrici e oltre.
